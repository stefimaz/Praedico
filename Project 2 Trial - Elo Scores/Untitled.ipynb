```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
nfl_elo_df = pd.read_csv(Path('Resources/nfl_elo.csv'), index_col='date', parse_dates = True, infer_datetime_format=True)

# Review the DataFrame
nfl_elo_df.head()
```

```python
# Selection of just certain features to try and test a potential model

nfl_2016_2021_df = nfl_elo_df.loc['2016-09-08':'2021-02-07'][['team1', 'team2', 'elo1_pre', 'elo2_pre', 'elo_prob1', 'elo_prob2', 'score1', 'score2']]
```

```python
nfl_2016_2021_df.head()
```

```python
# Create an ouptut variable: FavoredWin
# This variable will be a 1 if the favored team wins and a 0 if the favored team loses
# The favored team is that which has the higher Elo score pre-game

favored_win = []

for i in range(len(nfl_2016_2021_df)):
    if (nfl_2016_2021_df.loc[i, 'elo1_pre'] > nfl_2016_2021_df.loc[i, 'elo2_pre']) and (nfl_2016_2021_df.loc[i, 'score1'] > nfl_2016_2021_df.loc[i, 'score2']):
        favored_win.append('1')
    elif (nfl_2016_2021_df.loc[i, 'elo2_pre'] > nfl_2016_2021_df.loc[i, 'elo1_pre']) and (nfl_2016_2021_df.loc[i, 'score2'] > nfl_2016_2021_df.loc[i, 'score1']):
        favored_win.append('1')
    else:
        favored_win.append('0')
```
